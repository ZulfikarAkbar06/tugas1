```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
# CLEANING & PREPARATION ‚Äì ONE CELL

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")
print("Dataset Loaded. Shape awal:", df.shape)
df_original = df.copy()

# Missing Values Handling
threshold = int(len(df) * 0.6)
df = df.dropna(axis=1, thresh=threshold)

num_cols_raw = df.select_dtypes(include=[np.number]).columns
for col in num_cols_raw:
    df[col] = df[col].fillna(df[col].median())

cat_cols_raw = df.select_dtypes(include=["object"]).columns
for col in cat_cols_raw:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values selesai ditangani.")

# Outlier Handling (IQR)
num_cols = []
for col in df.select_dtypes(include=[np.number]).columns:
    if col.lower() not in ["passengerid", "ticket", "cabin"]:
        num_cols.append(col)

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))

print("Outliers ditangani (IQR winsorizing).")

# Normalisasi (Min-Max Scaling)
for col in num_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    if min_val != max_val:
        df[col] = (df[col] - min_val) / (max_val - min_val)

print("Normalisasi selesai.")

# 5. Encoding kategorikal & hapus duplikat
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    df[col] = pd.factorize(df[col])[0]

df = df.drop_duplicates()
print("Encoding selesai & duplikat dihapus.")

# Visualisasi
num_cols_final = df.select_dtypes(include=[np.number]).columns

df[num_cols_final].hist(bins=20, figsize=(12, 8))
plt.suptitle("Distribusi Kolom Numerik Setelah Cleaning", fontsize=14)
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(df[num_cols_final].corr(), annot=True, cmap="coolwarm")
plt.title("Korelasi Antar Fitur Numerik")
plt.show()

# Summary akhir
print("\nUMMARY DATASET SETELAH CLEANING")
print("Shape akhir:", df.shape)
df.info()
print("\nCleaning selesai. Dataset siap digunakan untuk analisis!")
```
