```python
class Barang:
    def __init__(self, kode_barang: str, nama_barang: str, harga: float, jumlah_stok: int):
        self.kode_barang = kode_barang
        self.nama_barang = nama_barang
        self.harga = harga
        self.jumlah_stok = jumlah_stok

def update_stok(self, jumlah: int, tindakan: str):
    if jumlah < 0:
        raise ValueError("Jumlah stok tidak boleh negatif.")

    if tindakan == "tambah":
        self.jumlah_stok += jumlah
    elif tindakan == "kurang":
        if self.jumlah_stok < jumlah:
            raise ValueError("Jumlah stok tidak cukup untuk dikurangi.")
        self.jumlah_stok -= jumlah
    else:
        raise ValueError("Tindakan tidak valid. Gunakan 'tambah' atau 'kurang'.")


    def info_barang(self):
        return (f"Kode Barang: {self.kode_barang}\n"
                f"Nama Barang: {self.nama_barang}\n"
                f"Harga: {self.harga}\n"
                f"Jumlah Stok: {self.jumlah_stok}\n")



daftar_barang = []
kode_barang_dict = {}

def tambah_barang(kode_barang, nama_barang, harga, jumlah_stok):
    if kode_barang in kode_barang_dict:
        raise ValueError("Kode barang sudah ada.")
    barang_baru = Barang(kode_barang, nama_barang, harga, jumlah_stok)
    daftar_barang.append(barang_baru)
    kode_barang_dict[kode_barang] = barang_baru

def tampilkan_inventaris():
    if not daftar_barang:
        print("Inventaris kosong.")
    else:
        print("Inventaris Barang:")
        for barang in daftar_barang:
            print(barang.info_barang())

def cari_barang(kode: str):
    if kode in kode_barang_dict:
        print("Barang Ditemukan:")
        print(kode_barang_dict[kode].info_barang())
    else:
        print("Barang dengan kode tersebut tidak ditemukan.")


print("---------MANAJEMEN INVENTARIS BARANG----------")
print("______________________________________________")
print("|                                            |")
print("|        1. Tambah barang baru               |")
print("|        2. Update jumlah stok               |")
print("|        3. Cari barang                      |")
print("|        4. Tampilkan inventaris             |")
print("|____________________________________________|")
print("")

while True:
    try:
        pilihan = int(input("Masukkan pilihan Anda (1/2/3/4): "))
        if pilihan == 1:
            try:
                kode_barang_baru = input("Masukkan kode barang: ").strip()
                nama_barang_baru = input("Masukkan nama barang: ").strip()
                harga_baru = float(input("Masukkan harga barang: "))
                jumlah_stok_baru = int(input("Masukkan jumlah stok: "))
                tambah_barang(kode_barang_baru, nama_barang_baru, harga_baru, jumlah_stok_baru)
                print("Barang berhasil ditambahkan.")
            except ValueError as e:
                print(f"Kesalahan: {e}")

        elif pilihan == 2:
            try:
                kode_barang = input("Masukkan kode barang: ").strip()
                if kode_barang not in kode_barang_dict:
                    raise KeyError("Kode barang tidak ditemukan.")
                tindakan = input("Tambah/Kurang: ").strip().lower()
                if tindakan not in ["tambah", "kurang"]:
                    raise ValueError("Tindakan tidak valid. Gunakan 'tambah' atau 'kurang'.")
                jumlah_stok_baru = int(input("Masukkan jumlah stok: "))
                kode_barang_dict[kode_barang].update_stok(jumlah_stok_baru, tindakan)
                print("Jumlah stok berhasil diupdate.")
            except ValueError as e:
                print(f"Kesalahan: {e}")
            except KeyError as e:
                print(f"Kesalahan: {e}")

        elif pilihan == 3:
            kode_barang_cari = input("Masukkan kode barang: ").strip()
            cari_barang(kode_barang_cari)

        elif pilihan == 4:
            tampilkan_inventaris()

        else:
            print("Pilihan tidak valid.")
    except ValueError:
        print("Masukkan pilihan dalam angka.")

```
